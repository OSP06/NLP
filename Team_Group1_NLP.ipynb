**String Manipulation**

```python
def delete_char(word):
  return [word[:i] + word[i+1:] for i in range(len(word))]
print(delete_char('university'))
print(len(delete_char('university')))
```

```python
def switch_char(word):
  return [word[:i] + word[i+1] + word[i] + word[i+2:] for i in range(len(word) - 1)]
print(switch_char('Google'))
print(len(switch_char('Google')))
```

```python
from os import replace
import string

def replace_char(word):
  alpha = string.ascii_lowercase
  return [word[:i] + b + word[i+1:] for i in range(len(word)) for b in alpha if word[i] !=b]
print(replace_char('bat'))
print(len(replace_char('bat')))
```

```python
def insert_char(word):
  alpha = string.ascii_lowercase
  return [word[:i] + c + word[i:] for i in range(len(word) + 1) for c in alpha]   # iterates through all possible positions including the position after last character.
print(insert_char('ace'))
print(len(insert_char('ace')))
```

**Edit Distance Calculations**


```python
def edit_distance_one(word, allow_switches = True):
  #single edits with pervious functions
  deletes = delete_char(word)
  inserts = insert_char(word)
  replaces = replace_char(word)

  switches = switch_char(word) if allow_switches else []

  return set(deletes + inserts + replaces + switches)

print(edit_distance_one('ar', allow_switches=True))
print(len(edit_distance_one('ar', allow_switches=True)))
```

```python
def edit_distance_two(word, allow_switches=True):
  one_edit_words = edit_distance_one(word, allow_switches)

  two_edit_words = set()
  for w in one_edit_words:
    two_edit_words.update(edit_distance_one(w, allow_switches))
  return two_edit_words

two_edit_results = edit_distance_two('ace', allow_switches=True)
print(f"First 13 strings: {sorted(list(two_edit_results))[:13]}")
print(f"Last 13 strings: {sorted(list(two_edit_results))[-13:]}")
print(f"{len(two_edit_results)}")

```

**Edit Distances with Spelling Suggestions**

```python
from collections import Counter, defaultdict
import re

def Read(filepath):
    with open(filepath, 'r') as file:
        text = file.read()
    return set(text.lower().split())  # Convert to lowercase and split into words

def edit_distance_one(word, allow_switches=True):
    deletes = delete_char(word)
    inserts = insert_char(word)
    replaces = replace_char(word)
    switches = switch_char(word) if allow_switches else []
    return set(deletes + inserts + replaces + switches)

def edit_distance_two(word, allow_switches=True):
    return set(e2 for e1 in edit_distance_one(word, allow_switches)
               for e2 in edit_distance_one(e1, allow_switches))

def fix_edits(word, vocab, allow_switches=True, n=5):
    word = word.lower()  #converted to lowercase to ensure case-insensitive matching with the vocabulary.
    suggestions = set()

    if word in vocab:
        return [(word, 1.0)]
# generate all possible words that are one edit away from the input word. filtered by checking which of these edits are words
    one_edits = edit_distance_one(word, allow_switches)
    valid_one_edits = one_edits.intersection(vocab)
# if no valid one-edit suggestions found, the function proceeds to generate suggestions that are two edits
    if valid_one_edits:
        suggestions.update(valid_one_edits)

    if not suggestions:
        two_edits = edit_distance_two(word, allow_switches)
        valid_two_edits = two_edits.intersection(vocab)
        suggestions.update(valid_two_edits)

    if not suggestions:
        return [(word, 0.0)]
# the function calculates the probability of each suggestion.
    word_counts = defaultdict(lambda: 1, Counter(vocab))
    total_words = sum(word_counts.values())

    best_words = [(s, word_counts[s] / total_words) for s in suggestions]
    best_words.sort(key=lambda x: -x[1])  # Sort by probability in descending order

    return best_words[:n]

def delete_char(word):
    return [word[:i] + word[i+1:] for i in range(len(word))]

def switch_char(word):
    return [word[:i] + word[i+1] + word[i] + word[i+2:] for i in range(len(word) - 1)]

def replace_char(word):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    return [word[:i] + c + word[i+1:] for i in range(len(word)) for c in alphabet if word[i] != c]

def insert_char(word):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    return [word[:i] + c + word[i:] for i in range(len(word) + 1) for c in alphabet]

def find_misspelled_words(text_file, vocab_file):
    # Read the vocabulary
    vocab = Read(vocab_file)

    # Read the text file and extract words
    with open(text_file, 'r') as file:
        text = file.read().lower()
    words = re.findall(r"\b[\w'.,!?;:]+\b", text)  # Regular expression to match words

    # Find misspelled words
    misspelled = [word for word in words if word not in vocab]

    # Get suggestions for misspelled words
    suggestions = {word: fix_edits(word, vocab) for word in misspelled}

    return suggestions

# Example usage
vocab_file = '/content/THE SONNETS.txt' # sample vocabulary list
text_file = '/content/shakespeare.txt'  # text file path

misspelled_words = find_misspelled_words(text_file, vocab_file)

if misspelled_words:
    print("Misspelled words found:")
    for word, suggs in misspelled_words.items():
        print(f"\nMisspelled word: {word}")
        print(f"Suggestions: {[word for word, _ in suggs]}")
        for i, (sugg, prob) in enumerate(suggs, 1):
            print(f"  Suggestion {i}: {sugg}, probability {prob:.6f}")
else:
    print("No misspelled words found.")
print(len(misspelled_words))
```

```python
from collections import Counter
import string

# Function to read the vocabulary from the Shakespeare text file
def your_read_function(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        # Read all words, convert to lowercase and remove punctuation
        text = file.read().lower()
        words = set(re.findall(r"\b[\w']+\b", text))  # We use a set to store unique words
    return words

# Redefining the necessary edit functions
def delete_char(word):
    return [word[:i] + word[i+1:] for i in range(len(word))]

def switch_char(word):
    return [word[:i] + word[i+1] + word[i] + word[i+2:] for i in range(len(word) - 1)]

def replace_char(word):
    alphabet = string.ascii_lowercase
    return [word[:i] + c + word[i+1:] for i in range(len(word)) for c in alphabet if word[i] != c]

def insert_char(word):
    alphabet = string.ascii_lowercase
    return [word[:i] + c + word[i:] for i in range(len(word) + 1) for c in alphabet]

def edit_distance_one(word, allow_switches=True):
    deletes = delete_char(word)
    inserts = insert_char(word)
    replaces = replace_char(word)
    switches = switch_char(word) if allow_switches else []
    return set(deletes + inserts + replaces + switches)

def edit_distance_two(word, allow_switches=True):
    one_edit_words = edit_distance_one(word, allow_switches)
    two_edit_words = set()
    for w in one_edit_words:
        two_edit_words.update(edit_distance_one(w, allow_switches))
    return two_edit_words

# Function to suggest the best words
def fix_edits(word, vocab, n=5, allow_switches=True):
    # Step 1: Check if the word is already in the vocabulary
    if word in vocab:
        return [(word, 1.0)]  # The word is perfectly valid, return with full probability

    # Step 2: Get suggestions from edit_distance_one
    one_edits = edit_distance_one(word, allow_switches)
    one_edit_suggestions = one_edits.intersection(vocab)

    if one_edit_suggestions:
        # Count frequencies (probabilities) of one-edit suggestions
        best_words = Counter({w: 1 / len(vocab) for w in one_edit_suggestions})
        return best_words.most_common(n)

    # Step 3: Get suggestions from edit_distance_two if no one-edit suggestions are found
    two_edits = edit_distance_two(word, allow_switches)
    two_edit_suggestions = two_edits.intersection(vocab)

    if two_edit_suggestions:
        # Count frequencies (probabilities) of two-edit suggestions
        best_words = Counter({w: 1 / len(vocab) for w in two_edit_suggestions})
        return best_words.most_common(n)

    # Step 4: If no suggestions, return the input word with probability 0
    return [(word, 0.0)]

# Filepath to shakespeare.txt
file_path = '/content/shakespeare.txt'

# Reading the vocabulary from the file
vocab = your_read_function(file_path)

# Example usage with the word
print(fix_edits('dys', vocab, n=5))
print(len(fix_edits('dys', vocab, n=5)))
```
